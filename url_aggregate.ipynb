```python
import requests
import json
import pandas as pd
import numpy as np
import datetime as dt
```

```python
url = 'https://stats.nba.com/stats/leaguegamelog?Counter=1000&DateFrom=&DateTo=&Direction=DESC&ISTRound=&LeagueID=00&PlayerOrTeam=T&Season=2022-23&SeasonType=Regular%20Season&Sorter=DATE'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64)', 'x-nba-stats-origin': 'stats', 'x-nba-stats-token': 'true', 'Host':'stats.nba.com', 'Referer':f'https://www.nba.com/'}
```

```python
r= requests.get(url, headers=headers, timeout = 7)
data = json.loads(r.text)
```

```python
games = data['resultSets'][0]['rowSet']
```

```python
ids = [[info[4],info[6].split(' @ ')[0],info[6].split(' @ ')[1]] for info in games if '@' in info[6]]
ids = sorted(ids)
ids
```

```python
df = pd.DataFrame.from_dict(ids)
df.columns=[['game_id','home','away']]
```

```python
df.to_csv('game_ids.csv',index=False)
```
